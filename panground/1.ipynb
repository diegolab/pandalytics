O método info() no Pandas é usado para obter um resumo conciso de um DataFrame, incluindo informações sobre o tipo de índice de linhas e colunas, os valores não nulos e a memória usada.

As informações fornecidas pelo método info() podem ser úteis para:

Validar a estrutura de dados do DataFrame. Por exemplo, podemos verificar se todas as colunas têm o mesmo comprimento, ou se não há valores ausentes em colunas que não deveriam ter valores ausentes.
Obter uma visão geral rápida dos dados do DataFrame. Por exemplo, podemos verificar quais são os tipos de dados das colunas, ou qual é a distribuição dos valores em cada coluna.
O método info() pode ser usado com qualquer tipo de DataFrame, incluindo DataFrames criados a partir de arquivos CSV, JSON, Excel ou outros formatos de dados.

```python
titanic.describe()
```

O método describe() no Pandas é usado para obter um resumo estatístico de um DataFrame, incluindo informações sobre a distribuição dos dados em cada coluna.

A saída do método describe() é um objeto DataFrame com as seguintes colunas:

count: O número de valores não nulos em cada coluna.
mean: A média dos valores em cada coluna.
std: O desvio padrão dos valores em cada coluna.
min: O valor mínimo em cada coluna.
25%: O primeiro quartil dos valores em cada coluna.
50%: A mediana dos valores em cada coluna.
75%: O terceiro quartil dos valores em cada coluna.
max: O valor máximo em cada coluna.

As informações fornecidas pelo método describe() podem ser úteis para:

Obter uma visão geral rápida dos dados do DataFrame. Por exemplo, podemos verificar quais são os valores centrais da distribuição, como a média, a mediana e o desvio padrão.
Identificar outliers. Outliers são valores que estão muito distantes da maioria dos outros valores na distribuição. Podemos identificar outliers observando os valores máximos e mínimos da distribuição.
Comparar diferentes colunas do DataFrame. Podemos comparar as distribuições de diferentes colunas observando as estatísticas descritivas de cada coluna.
O método describe() pode ser usado com qualquer tipo de DataFrame, incluindo DataFrames criados a partir de arquivos CSV, JSON, Excel ou outros formatos de dados.

```python
titanic.describe(include="O")
```

O parâmetro include do método describe() permite especificar quais estatísticas descritivas serão calculadas para cada coluna. O valor padrão do parâmetro include é "all", que significa que todas as estatísticas descritivas serão calculadas.

O parâmetro include também pode ser usado para excluir estatísticas descritivas específicas. Por exemplo, o seguinte código exclui as estatísticas min, 25% e max:

import pandas as pd

df = pd.DataFrame({
    "nome": ["João", "Maria", "Pedro"],
    "idade": [20, 25, 30],
    "altura": [1.80, 1.70, 1.60],
    "sexo": ["M", "F", "M"]
})

df.describe(include=["count", "mean", "std", "median"])

O parâmetro include pode ser usado para especificar uma lista de estatísticas descritivas específicas que devem ser calculadas. Por exemplo, o seguinte código especifica que apenas as estatísticas count, mean e std devem ser calculadas:
